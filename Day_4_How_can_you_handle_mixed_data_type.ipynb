```python
# Now we have to work on the column which have both numerical and categorical in a simgle cell of the column 
df1=pd.DataFrame(data["Ticket"])
```

```python
df1.head(15) '''Now you can look that in this dataset their is one oattern that after the categorical value their is the gap and after that 
numerical value '''
```

```python
# Now in This we have to split this data into numerical in one column and categerical type of data in another column 
df1["Ticket_numerical"]=df1["Ticket"].str.extract('(\d+)$')
```

```python
df1["Ticket_categorical"] = df1["Ticket"].str.extract(r'^(.+?)\s+\d+$')

df1["Ticket_categorical"]
```

```python
df1
```

```python
# Now can also fill the null values with missing in categorical and 0 in Numerical 
df1["Ticket_categorical"]=df1["Ticket_categorical"].fillna('Missing')
```

```python
# Now can also fill the null values with missing in categorical and 0 in Numerical 
df1["Ticket_numerical"]=df1["Ticket_numerical"].fillna('0')
```
